```python
import os
import shutil
import re 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np
import cv2
```

```python
# first copying one sample image from every folder to a single folder
img_folders = os.listdir('train')
base_path = "train/"

for img_folder in img_folders:
    img_folder_path = base_path + img_folder + '/'
    img_list = os.listdir(img_folder_path)
    file = img_list[5]
    img_path = img_folder_path + file
    shutil.copy2(img_path, 'collage_sample/{}'.format(file))

    
```

```python
file_list = os.listdir('collage_sample')

images = []
for file in file_list:
    img = cv2.imread('collage_sample/' + file)
    res = cv2.resize(img, dsize=(28,28))
    res = np.invert(res)
    images.append(res)

fig = plt.figure(1, (30, 30))
grid = ImageGrid(fig, 110,
                 nrows_ncols=(9, 10),
                 axes_pad=.5,
)



clean_file_list = [re.sub('[0-9]+.png', '', s) for s in file_list]

for i in range(len(images) + 4):
    try: # to get rid of empty subplot on bottom right
        grid[i].imshow(images[i])
        grid[i].set_title(clean_file_list[i], size = 25)
    except:pass
    
    grid[i].axis('off')


plt.show(block=True);

fig.savefig('collage_sketch.jpg');
```
